```python
import numpy as np 
import matplotlib.pyplot as pyplot 


data = np.loadtxt('data.txt')
x = data[:, 0]
y1 = data[:, 1]
y2 = data[:, 2]

xx = [1024, 2048, 3072, 4096, 5120, 6144, 7168, 8192, 9216, 10240 ]
# print(np.max(y1/y2))
print(y1/y2)

pyplot.plot(x, y1, label = "Kernel 8", color = "coral")
pyplot.plot(x, y2, label = "cuBLAS", color = "blue")
pyplot.plot(x, y1, 'o', color = "darkred")
pyplot.plot(x, y2, 'P', color = "navy")
# pyplot.ylim(0.0, 80)
pyplot.ylabel("GFLOPS")
pyplot.xlabel("Matrix Size M=N=K")
pyplot.xticks(xx)
pyplot.legend()
pyplot.savefig("flops.png", dpi = 200)
pyplot.show()
```
